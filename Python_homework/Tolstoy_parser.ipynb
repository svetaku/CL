```
! pip3 install requests
! pip3 install beautifulsoup4

import requests as rq

from bs4 import BeautifulSoup
```

```
url = "http://az.lib.ru/t/tolstoj_lew_nikolaewich/"
page = rq.get(url)
page
```

```
soup_books = BeautifulSoup(page.text, 'html.parser')
```

```
url_book_list = []
ind = 0

for i in soup_books.find_all('dt'):
      book_url = url + i.find_all("a")[0]["href"]
      url_book_list.append(book_url)

url_book_list    #получаем грязный список ссылок со страницы автора
```

```
url_book_list1 = url_book_list.copy()
final_ind = url_book_list.index('http://az.lib.ru/t/tolstoj_lew_nikolaewich/text_1906_bojeskoe_i_chelovecheskoe.shtml')   # последняя ссылка в списке произведений, которая нас интересует

for x in url_book_list1:
      if "ch//" in x:
          url_book_list.remove(x)
      if "ch/http" in x:
          url_book_list.remove(x)
      try:
        if "text_0073.shtml" in x:            # первый вариант романа "Война и мир" нам не нужен
          url_book_list.remove(x)
      except:
          continue
      try:
         if url_book_list.index(x) >= final_ind:
           url_book_list.remove(x)
      except:
          continue    

url_book_list   #получаем финальный список ссылок на произведения
```

```
book_corp = ''
for book in url_book_list:
      page_book = rq.get(book)
      soup_book = BeautifulSoup(page_book.text, 'html.parser')
      try:
         book_text = soup_book.find("div", align="justify").text
         print(book)
         # достаем тексты из ссылок и печатаем отработанные ссылки
      except: 
         try: 
           book_text = soup_book.find("pre").text     # страницы с разными произведениями имеют разную html-структуру, поэтому у цикла две части - на оба случая
           print(book)
         except:
            continue
      book_corp += book_text
```

```
f = open('tolstoy_corp1.txt', 'w')
f.write(book_corp)
f.close()
```
